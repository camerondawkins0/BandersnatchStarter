```python
import altair as alt

from altair import Chart, X, Y, Color, Tooltip, Axis
from pandas import DataFrame
```

```python
from data import Database
from graph import chart
db = Database()

chart_selection_interval = chart(db.dataframe(), x='Health', y='Energy', target='Rarity')

chart_selection_interval
```

```python
graph_hist = Chart(db.dataframe()).mark_bar().encode(x=X(field='Level',
                                                         type='quantitative',
                                                         bin=alt.BinParams(maxbins=15)),
                                                     y=Y(field='Type', aggregate='count'))

graph_hist
```

```python
graph_boxplot = Chart(db.dataframe()).mark_boxplot(extent='min-max').encode(x='Type:O', y='Health:Q')
graph_boxplot
```
